# Generalization of Welding Machine with Edge Detection and Precise Control


## Abstract

This project presents an innovative approach to welding automation, tailored for precision welding in confined spaces. Inspired by 3D printer design principles, our system eliminates the need for Z-axis movement, ensuring stability and adaptability for welding tasks. High-precision linear rails and pulley systems facilitate smooth and accurate movement.

A key feature of our system is real-time monitoring and control. We integrated a camera for live monitoring and quality control during welding, while a laser module visually represents the welding process, allowing for immediate feedback and adjustments. Control is achieved through an Arduino Uno microcontroller with a CNC Shield, which precisely controls Nema 17 Stepping Motors using A4988 Stepper Motor Drivers.

Calibration was performed meticulously through extensive trial and testing, guaranteeing optimal alignment and performance. Our system's adaptability and precision make it suitable for various industries, including automotive, aerospace, and electronics manufacturing. It offers cost-efficiency, a user-friendly interface, safety measures, and future readiness for further technological advancements.

This project showcases our expertise in welding automation, component selection, and calibration, as well as our commitment to safety and usability. It stands as a testament to effective teamwork and sets the stage for future developments in welding automation technology.
